## What is Pandas
Pandas is a popular open-source data manipulation and analysis library for Python. It provides data structures for efficiently storing and manipulating large datasets and tools for working with structured data. One of the major reasons we use pandas is because it requires less laborous work as compared to working manually with sheets. It also allows for automation and makes it very easy when working with big data
# DataFrame
A two-dimensional table with labeled axes (rows and columns). It is similar to a spreadsheet or a table and is one of the most commonly used Pandas data structures.

# Series
A one-dimensional labeled array capable of holding any data type. It is similar to a column in a DataFrame or a single-column table.

# null values
a dataframe might contain null or missing values
these values cause alot of problems when performing mathematical operations

```python
# here is how you can check if a df contains null values
df.isna()

# this gives the sum of all null values present in the df
df.isna().sum()
```

# removing null values:
there are various ways to remove null values in a df,
we will discuss three

```python
# fillna has alot of options, we will discuss three: 'bfill', 'ffill', 'mean'

# ffill replaces the value with the anteceding value in the same column
df['example_column'].fillna(method = 'ffill', inplace = True)

# bfill replaces the value with the preceding value in the same column
df['example_column'].fillna(method = 'bfill', inplace = True)

# sometimes we require null values to be filled by the mean

# here is how to calculate the mean:
mean = df['example_column'].mean()
# we will then fill the nan values with the mean
df['example_column'].fillna(mean, inplace = True)

# checking for null vals
df.isna().sum()
```

# -> data visualization
we can visualize data using matplotlib, a popular data visualization library in python. It offers a multitude of 2D and 3D visualization functions

```python
# import the library from system | assigning it to 'plt' makes it easier to type 
import matplotlib.pyplot as plt
```

# bar graph

```python
# bar graph (vertical)
short = df.head()

plt.bar(short['first_name'], short['biology_score'], color = 'r')
plt.xlabel('names')
plt.ylabel('score')
plt.title('biology score of students')
plt.show()
```

```python
# bar graph (horizontal)
short = df.head()

plt.barh(short['first_name'], short['biology_score'], color = 'b')
plt.xlabel('names')
plt.ylabel('score')
plt.title('biology score of students')
plt.show()
```
